# Apply: Se usa para aplicar a una columna o conjunto de ellas una funci√≥n creada por nosotros. Muy util

```python
import pandas as pd
import numpy as np
```

```python
path = '/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv'
df = pd.read_csv(path)
df.head(4)
```

```python
def two_times(value):
  return value * 2
```

```python
df['User Rating'].apply(two_times) # Esto es muchisimo mas eficiente que usar un ciclo for para igual tarea
```

```python
df['User Rating x 2'] = df['User Rating'].apply(two_times)
```

```python
df.head()
```

**El apply tmb se puede usar con funciones lambda lo que lo hace extremadamente versatil**

```python
df['User Rating x 3'] = df['User Rating'].apply(lambda x : x *3) # Es mas eficiente a√∫n ya no guardo una funci√≥n en memoria
```

```python
df.head()
```

# Puedo aplicar tmb un apply a todo el dataframe o a distintas partes de el planteando condicionales. Por ejemplo: 

```python
df['User Rating x 2'] = df.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1) # Para que funcione los condicionales debo especificar a que eje hago referencia.
# En este caso al eje 1 de las columnas
```
