```python
# Drop duplicate : Elimina las filas duplicadas de un data set
df.iloc[0] # Miró mi primer registro
```

```python
df = df.append(df.iloc[0]) # Insertó al final mi primer registro para que esté duplicado. 
```

```python
df.drop_duplicates() # Elimina todos los duplicados de mis filas. 
```

```python
df.drop_duplicates(keep='last') # Va a eliminar el primero de los dos duplicados en lugar del ultimo. 
```
