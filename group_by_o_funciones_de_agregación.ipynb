```python
import pandas as pd
```

```python
path = '/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv'
df = pd.read_csv(path)
df
```

```python
df.groupby('Author').min() # Cuenta de cantidad de registros por autor
```

```python
df.groupby('Author').mean()
```

```python
df.groupby('Author').count()
```

```python
df.groupby('Author').sum()
# Autor en este momento no es una columna sino que es el indice de mi dataframa por lo que podría filtralo por label con loc
```

```python
df.groupby('Author').sum().loc['William Davis']
```

```python
# Si no quiero que los autores sean el indice y quiero usar un indice numerico debo resetear el indice en mi group by
df.groupby('Author').sum().reset_index()
```

```python
# Puedo pedirle mas de una función de agregación para un mismo group by con la función .agg
df.groupby('Author').agg(['min','max'])
```

```python
# Con UN diccionarios tmb puedo pedirle distintas medidas de agregación para distintas columnas
df.groupby('Author').agg({'Reviews':['min','max'] , 'Price' : 'sum'})
```

```python
# tmb puedo agrupar o agregar por mas de una variable
df.groupby(['Author','Year']).count()
```

```python
df.groupby(['Author','Year']).count().reset_index()
```
