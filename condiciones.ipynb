```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df.sample(5)
```

```python
# Condiciones. Muy similar a Numpy
df['Year'] > 2016 # Me devuelve con True aquellas rows que cumplan la condición
```

```python
df[df['Year'] > 2016] # Me devuelve el dataset filtrado por aquellos rows que cumplen la condición 
```

```python
# Pandas está escrito sobre numpy. Por lo que muchas funciones se parecen 
df[(df['Year'] > 2016) & (df['Author'] == 'Jen Sincero')] # Filtrado por multiples condiciones. 
```

```python
# Tmb podría guardar las condiciones en una variable y filtrar por esas variables
conditions = (df['Year'] > 2016) & (df['Author'] == 'Jen Sincero')
df[conditions]
```

```python
# En pandas puedo negar la condicion si recurrir a construir otro condicional
mayor_a_2016 = df['Year'] > 2016
df[~mayor_a_2016] # Niego la condicion y busco los menores o iguales al 2016
```
