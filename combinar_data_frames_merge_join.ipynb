#  Merge

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
der
```

```python
izq
```

```python
izq.merge(der, on='key') # Podría no especificar columna de merge porque pandas la detecta automaticamente pero no es recomedable por la claridad del codigo. 
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
# Si los nombres de las columnas no coinciden pero si lo hacen sus valores entonces puedo especificar sus nombre en el codigo y mergearlo igual
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2') # No me trae 'k3' porque esta haciendo un inner join
```

```python
# Que tipo de combinación hacer puede ser seteado mediante el parametro how dentro del merge
izq.merge(der, left_on='key', right_on='key_2', how = 'inner') # Es como lo hace por default
```

```python
izq.merge(der, left_on='key', right_on='key_2', how = 'left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how = 'right')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how = 'outer') # con outer hago el merge completo rellenando de NaN los que no mergearon
```

# Join: Index match vs match de columnas especificas como hicimos con el merge

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3']) 
```

```python
izq # Estos dataframes tienen indices seteados de antemano
```

```python
der
```

```python
# El join entonces va a servir para unir dataframes cuando tengamos indices seteados de antemano. No con los standar. Para estos casos debemos usar el merge y especificar que union realizar. 
izq.join(der) # Por defecto join será un left join. Si quiero hacer un inner, right o outer debo especificarlo como saegundo parametro
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='outer')
```
