```python
import pandas as pd
import numpy as np
```

```python
dict = {'Col1':[1,2,3,np.nan],
'Col2':[4, np.nan,6,7],
'Col3':['a','b','c', None]}
```

```python
dict
```

```python
df = pd.DataFrame(dict)
df
```

# ¿Como trabajamos en Pandas con los valores nulos o Nan?

```python
df.isnull() # Si el valor es NaN o None entonces nos dará True
```

```python
df.isnull()*1  # Si el valor es NaN o None entonces nos dará 1. Sirve mucho para redes neuronales
```

```python
# Reemplazo los valores nulos por la palabra seleccionada. En este caso Missing
df.fillna('Missing')
```

```python
# Tmb podría reemplazar los nulos por la media o cualquier otro valor estadistico siempre que mi columna sea de numeros. 
df.fillna(df.mean())
```

```python
# Otra forma de llenar los valores nulos es df.interpolate()
df.interpolate() # Rellena los espacios intermedios con los valores que Pandas asume deberían ir en esa serie. Si es el minimo o el maximo los repite. 
```

```python
# Otra manera que muchos cientificos usan es borrar directamente las filas que traigan datos nulos
df.dropna()
```
